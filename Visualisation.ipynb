```python
    import matplotlib.pyplot as plt
    import numpy as np
    import math
    from mpl_toolkits.mplot3d import Axes3D
    
    n = 128
    x = [i/n for i in range(0,n+1)]
    y = [i/n for i in range(0,n+1)]
    z = [[math.sin(math.pi*i/n)*math.sin(math.pi*j/n) for i in range(0,n+1)] for j in range(0,n+1)]

    f = open("out.txt")
    r = f.read().split(' ')
    f.close()
    
    a = [[0. for i in range(0,n+1)] for j in range(0,n+1)]
    for i in range (0, n+1):
        for j in range (0, n+1):
            a[i][j] = float(r[i*(n+1)+j])
    
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.contour3D(x, y, z, 50, cmap='inferno')
    plt.show()
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.contour3D(x, y, a, 50, cmap='inferno')
    plt.show()    
```
